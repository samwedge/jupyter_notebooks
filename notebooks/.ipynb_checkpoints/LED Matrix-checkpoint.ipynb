```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
width = 4
height = 4
```

```python
def plot_led_matrix():
    f = plt.figure(figsize=(5, 5))
    for i in range(height):
        for j in range(width):
            x = xx[i, j]
            y = yy[i, j]
            r = r_grid[i, j]
            g = g_grid[i, j]
            b = b_grid[i, j]
            plt.plot(x, y, 'o', color=(r, g, b, 1), figure=f, markersize=30)
    plt.axis('equal')
    ax = f.axes[0]
    ax.set_aspect('equal', 'box')
    ax.set_facecolor((0, 0, 0))
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())
    plt.xlim(-1, width)
    plt.ylim(-1, height)
    plt.show()
```

```python
def point(x, y, strength=10):
    grid = np.zeros((height, width))
    for i in range(height):
        for j in range(width):
            d = np.sqrt((xx[i, j] - x) ** 2 + (yy[i, j] - y) ** 2)
            if d > strength:
                grid[i, j] = 0
            elif d == 0:
                grid[i, j] = 1
            else:
                grid[i, j] = 1 - d / strength
    return grid
```

```python
for a in range(5):
    r_grid = point(a, 0, 5)
    g_grid = point(a, a, 5)
    b_grid = point(a, 2, 5)
    xx, yy = np.meshgrid(range(width), range(height))
    plot_led_matrix()
    time.sleep(0.5)
```
