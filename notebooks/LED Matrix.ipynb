```python
import time
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
width = 4
height = 4
xx, yy = np.meshgrid(range(width), range(height))
```

```python
%%capture
fig, ax = plt.subplots(figsize=(8, 8))

ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

sc = ax.scatter(x=[], y=[], s=1500)

ax.set_aspect('equal', 'box')
ax.set_facecolor((0, 0, 0))
ax.xaxis.set_major_locator(plt.NullLocator())
ax.yaxis.set_major_locator(plt.NullLocator())
ax.axis('equal')

plt.xlim(-1, width)
plt.ylim(-1, height)
```

```python
def point(x, y, strength):
    grid = 1 - np.sqrt((xx - x) ** 2 + (yy - y) ** 2) / strength
    grid[grid > 1] = 1
    grid[grid < 0] = 0
    return grid
```

```python
def animate(i):
    x = 5 / frames * i
    r_grid = point(x, 2, 2).ravel()
    g_grid = point(x, x, 2).ravel()
    b_grid = point(2, x, 2).ravel()
    
    x_arr = xx.ravel()
    y_arr = yy.ravel()
    c_arr = np.c_[r_grid, g_grid, b_grid]
    
    sc.set_offsets(np.c_[x_arr, y_arr])
    sc.set_color(c_arr)
```

```python
frames = 100
anim = animation.FuncAnimation(fig, animate, frames=frames, interval=30)
```
